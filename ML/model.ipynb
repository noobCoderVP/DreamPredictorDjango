# Machine Learning model for dreampredictor 

## Importing important libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Reading the CSV file

```python
df = pd.read_csv('data.csv', sep='\t')
df.head()
```

## Plotting the graphs

```python
x = [i for i in range(df['final'].size)]

plt.plot(x, df['final'], c='r')
plt.plot(x, df['first'], c='b')
```

## Creating a model

```python
X = df[['first', 'second', 'third', 'fourth', 'fifth']]
Y = df['final']
regr = LinearRegression()
 
regr.fit(X, Y)
print(regr.score(X, Y))
```

## Predicting some values

```python
print(regr.predict(np.array([96, 11, 78, 18, 74]).reshape(1, -1)))
```
